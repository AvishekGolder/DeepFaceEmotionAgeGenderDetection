```python
import cv2
import matplotlib.pyplot as plt

img1 = cv2.imread("img1.jpg")
plt.imshow(img1)
plt.show()
obj1 = DeepFace.analyze(img1, actions=['age', 'gender', 'race', 'emotion'])
print("Your age is", obj1["age"])
print("Your gender is", obj1["gender"])
print("Your race is", obj1["race"])
print("Your emotion is", obj1["emotion"])
```
